# Label Encoding for Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df1=df.copy()
df1.head()
```

```python
c=['X0','X1','X2','X3','X4','X5','X6','X8']
for i in c:
    df1[i]=le.fit_transform(df1[i])
df1.head()
```

```python
df1.corr()
```

# Standardizing the Variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
df_scaled=scale.fit_transform(df1)
df_scaled=pd.DataFrame(df_scaled,columns=df1.columns)
df_scaled.head()
```

```python
x=df_scaled.drop(['ID','y'],axis=1)
y=df_scaled['y']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=123)
```

# XGBoost

```python
from sklearn.metrics import r2_score,mean_squared_error
from  xgboost import XGBRegressor
from math import sqrt
```

```python
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
xgb_y_pred=xgb.predict(x_test)
```

```python
r2=r2_score(y_test,xgb_y_pred)
r2
```

```python
r2_train=r2_score(y_train,xgb.predict(x_train))
r2_train
```

```python
rmse_test=sqrt(mean_squared_error(y_test,xgb_y_pred))
rmse_test
```

```python
rmse_train=sqrt(mean_squared_error(y_train,xgb.predict(x_train)))
rmse_train
```

# XGBoost with PCA

```python
from sklearn.decomposition import PCA
```

```python
sklearn_PCA=PCA(n_components=50)
```

```python
sklearn_PCA.fit(x_train)
```

```python
x_train_transformed=sklearn_PCA.transform(x_train)
x_test_transformed=sklearn_PCA.transform(x_test)
```

```python
xgb.fit(x_train_transformed,y_train)
```

```python
y_pred1=xgb.predict(x_test_transformed)
```

```python
r2_score(y_test,y_pred1)
```

```python
r2_score(y_train,xgb.predict(x_train_transformed))
```

# XGBoost with PCA and Gridsearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'learning_rate':[.1,.2,.3],'n_estimators':[50,100,200,300,500],'max_depth':[2,3,4,5,6]}
```

```python
grid_search=GridSearchCV(estimator=xgb,param_grid=param_grid,n_jobs=-1)
```

```python
grid_search.fit(x_train_transformed,y_train)
```

```python
y_grid=grid_search.predict(x_test_transformed)
```

```python
print('Test accuracy:',r2_score(y_test,y_grid))
```

```python
print('Train accuracy:',r2_score(y_train,grid_search.predict(x_train_transformed)))
```
